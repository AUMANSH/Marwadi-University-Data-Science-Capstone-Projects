```python
import pandas as pd
data_location = 'SAU-EEZ-242-v48-0-old.csv'
df = pd.read_csv(data_location)
df.rename(columns = {"fish_name": "common_name", "country": "fishing_entity"}, inplace = True)
df.to_csv('SAU-EEZ-242-v48-0.csv', header=True, index=False)
df.to_parquet('SAU-EEZ-242-v48-0.parquet')
df.to_parquet('SAU-EEZ-242-v48-0.parquet')
```

```python
import pandas as pd

# Assuming df is your DataFrame containing the data
# Filter for the relevant conditions
filtered_df = df[(df['area_name'].str.contains('Fiji')) &
                 (df['fishing_entity'] == 'Fiji') &
                 (df['year'] > 2000)]

# Group by year and fishing entity and sum the landed_value
result = (filtered_df.groupby(['year', 'fishing_entity'])['landed_value']
          .sum()
          .reset_index())

# Cast the summed values to a decimal format (with 2 decimal places)
result['ValueEEZCatch'] = result['landed_value'].astype(float).round(2)

# Select relevant columns
final_result = result[['year', 'fishing_entity', 'ValueEEZCatch']]

# Sort by year
final_result = final_result.sort_values(by='year')

# Display the result
print(final_result)
```

```python
import pandas as pd

# Assuming df is your DataFrame containing the data
# Filter for the relevant conditions
filtered_df = df[((df['area_name'].str.contains('Fiji')) | (df['area_name'].isnull())) &
                  (df['fishing_entity'] == 'Fiji') &
                  (df['year'] > 2000)]

# Group by year and fishing entity and sum the landed_value
result = (filtered_df.groupby(['year', 'fishing_entity'])['landed_value']
          .sum()
          .reset_index())

# Cast the summed values to a decimal format (with 2 decimal places)
result['ValueEEZAndOpenSeasCatch'] = result['landed_value'].astype(float).round(2)

# Select relevant columns
final_result = result[['year', 'fishing_entity', 'ValueEEZAndOpenSeasCatch']]

# Sort by year
final_result = final_result.sort_values(by='year')

# Display the result
print(final_result)
```
