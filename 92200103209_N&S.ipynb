```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

LOAD THE IRIS DATASET

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
a= pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```python
df
```

```python
a
```

Normalization (Min-Max Scaling)

```python
scaler_norm = MinMaxScaler()
df_normalized = pd.DataFrame(scaler_norm.fit_transform(df), columns=df.columns)
a_normalized = pd.DataFrame(scaler_norm.fit_transform(a), columns=a.columns)
```

```python
df_normalized
```

```python
a_normalized
```

 Standardization (Z-score Normalization)

```python
scaler_std = StandardScaler()
df_standardized = pd.DataFrame(scaler_std.fit_transform(df), columns=df.columns)
a_standardized = pd.DataFrame(scaler_std.fit_transform(a), columns=a.columns)
```
